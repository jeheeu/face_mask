# data 폴더에 있는 .png를 데이터 분리하기

```python
import os
import random
import shutil

# 경로 정의
data_folder = 'data'
train_folder = os.path.join(data_folder, 'train')
val_folder = os.path.join(data_folder, 'val')
test_folder = os.path.join(data_folder, 'test')

# 디렉토리가 존재하지 않으면 생성
os.makedirs(train_folder, exist_ok=True)
os.makedirs(val_folder, exist_ok=True)
os.makedirs(test_folder, exist_ok=True)

# 데이터 폴더에서 모든 .png 파일 가져오기
png_files = [f for f in os.listdir(data_folder) if f.endswith('.png')]

# 파일 목록 섞기
random.shuffle(png_files)

# 각 세트에 대한 파일 수 계산
total_files = len(png_files)
train_count = int(total_files * 0.6)
val_count = int(total_files * 0.2)
test_count = total_files - train_count - val_count

# 파일 분할
train_files = png_files[:train_count]
val_files = png_files[train_count:train_count + val_count]
test_files = png_files[train_count + val_count:]

# 파일 이동 함수 정의
def move_files(file_list, destination_folder):
    for file_name in file_list:
        src_path = os.path.join(data_folder, file_name)
        dst_path = os.path.join(destination_folder, file_name)
        shutil.move(src_path, dst_path)

# 파일을 각 디렉토리로 이동
move_files(train_files, train_folder)
move_files(val_files, val_folder)
move_files(test_files, test_folder)

print(f"{len(train_files)} 개의 파일을 {train_folder}로 이동했습니다.")
print(f"{len(val_files)} 개의 파일을 {val_folder}로 이동했습니다.")
print(f"{len(test_files)} 개의 파일을 {test_folder}로 이동했습니다.")
```

# .png 데이터와 같은 이름은 가진 .txt를 같은 폴더에 옮기기

```python
# .png 파일을 기준으로 .txt 파일을 이동하는 함수
def move_txt_files(png_folder, destination_folder):
    # png_folder에서 .png 파일 목록 가져오기
    png_files = [f for f in os.listdir(png_folder) if f.endswith('.png')]
    
    # 각 .png 파일에 대해 .txt 파일 찾아서 이동
    for png_file in png_files:
        # .png 파일의 기본 이름 가져오기
        base_name = os.path.splitext(png_file)[0]
        
        # 해당 .png 파일에 대응하는 .txt 파일 경로 설정
        txt_src_path = os.path.join(png_folder, f'{base_name}.txt')
        txt_dst_path = os.path.join(destination_folder, f'{base_name}.txt')
        
        # .txt 파일이 존재하면 이동
        if os.path.exists(txt_src_path):
            shutil.move(txt_src_path, txt_dst_path)

# .png 파일을 기준으로 .txt 파일을 각 디렉토리로 이동
move_txt_files(train_folder, train_folder)
move_txt_files(val_folder, val_folder)
move_txt_files(test_folder, test_folder)
```

# 하이퍼파미터 튜닝
- 학습률은 모델이 얼마나 빠르게 업데이트 되는 지 정의
- hyp.yaml에 기본 학습률을 줄이기 위한 정의
- YOLOv5의 기본 하이퍼파라미터 설정을 참고하여 hyp.yaml 파일에 필요한 모든 하이퍼파라미터를 포함시켜야 함

```python
import yaml


# 하이퍼파라미터 설정
hyp_params = {
    'lr0': 0.01,  # initial learning rate
    'lrf': 0.2,  # final learning rate (lr0 * lrf)
    'momentum': 0.937,  # SGD momentum
    'weight_decay': 0.0005,  # optimizer weight decay
    'warmup_epochs': 3.0,  # warmup epochs
    'warmup_momentum': 0.8,  # warmup initial momentum
    'warmup_bias_lr': 0.1,  # warmup initial bias lr
    'box': 0.05,  # box loss gain
    'cls': 0.5,  # cls loss gain
    'cls_pw': 1.0,  # cls BCELoss positive_weight
    'obj': 1.0,  # obj loss gain (scale with pixels)
    'obj_pw': 1.0,  # obj BCELoss positive_weight
    'iou_t': 0.20,  # IoU training threshold
    'anchor_t': 4.0,  # anchor-multiple threshold
    'fl_gamma': 0.0,  # focal loss gamma (efficientDet default is gamma=1.5)
    'hsv_h': 0.015,  # image HSV-Hue augmentation (fraction)
    'hsv_s': 0.7,  # image HSV-Saturation augmentation (fraction)
    'hsv_v': 0.4,  # image HSV-Value augmentation (fraction)
    'degrees': 0.0,  # image rotation (+/- deg)
    'translate': 0.1,  # image translation (+/- fraction)
    'scale': 0.5,  # image scale (+/- gain)
    'shear': 0.0,  # image shear (+/- deg)
    'perspective': 0.0,  # image perspective (+/- fraction), range 0-0.001
    'flipud': 0.0,  # image flip up-down (probability)
    'fliplr': 0.5,  # image flip left-right (probability)
    'mosaic': 1.0,  # image mosaic (probability)
    'mixup': 0.0,  # image mixup (probability)
    'copy_paste': 0.0  # segment copy-paste (probability)
}


# hyp.yaml 파일 생성
with open('hyp.yaml', 'w') as file:
    yaml.dump(hyp_params, file, default_flow_style=False)

print("hyp.yaml 파일이 생성되었습니다.")
```
